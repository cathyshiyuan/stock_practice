```python
import pandas as pd
import warnings
warnings.simplefilter('ignore')
from datetime import datetime, timedelta
import collections
import numpy as np
import numpy.ma as ma
import pandas_datareader as pdr
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
import statsmodels.api as sm
from arch import arch_model
from arch.univariate import ARCH, GARCH
import seaborn
style.use('ggplot')
%matplotlib inline
```

```python
def download_data():
    tickers = pd.read_csv('tickers.csv', index_col = 0)
    tickers = tickers['tickers'].values.tolist()
    start = datetime(2005, 1, 1)
    end = datetime(2017, 3, 1)
    close = pd.DataFrame()
    volume = pd.DataFrame()
    for ticker in tickers:
        try:
            close[ticker] = pdr.DataReader(ticker, 'yahoo',start, end)['Adj Close']
            volume[ticker] = pdr.DataReader(ticker, 'yahoo',start, end)['Volume']
        except Exception as e:

            print e
    close.to_csv('sti_close.csv')
    volume.to_csv('sti_volume.csv')
```

```python
def cvt_index_dt(df):
    new_idx = []
    for idx in df.index:
        idx1 = datetime.strptime(idx,'%Y-%m-%d')
        new_idx.append(idx1)
    df.index = pd.to_datetime(df.index)
    return df

def xmas_time():
    christmas_start = []
    christmas_end = []
    for i in range(2005,2017):
        start = '-12-15'
        end = '-12-25'
        start = str(i) + start
        christmas_start.append(start)
        end = str(i) + end
        christmas_end.append(end)
    return christmas_start,christmas_end


def year():
    y_start = []
    y_end = []
    for i in range(2005,2017):
        start = '-01-01'
        end = '-12-31'
        start = str(i) + start
        end = str(i) + end
        y_start.append(start)
        y_end.append(end)
    return y_start, y_end


def strip_days(date, strip, goback = False):
    strip_date = datetime.strftime(datetime.strptime(date,'%Y-%m-%d') + timedelta(days =strip), '%Y-%m-%d')
    if goback == True:
        strip_date = datetime.strftime(datetime.strptime(date,'%Y-%m-%d') - timedelta(days =strip), '%Y-%m-%d')
    return strip_date


def xmas_return(with_year = True):
    xmas_return_df = pd.DataFrame()
    y_return_df = yearly_return()
    for i in range(len(xmas_time()[0])):      
        christmas_start = xmas_time()[0][i]
        christmas_end =  xmas_time()[1][i]
        
        xmas = close.ix[christmas_start: christmas_end]
          
        xmas_g = (xmas.shift(-1) - xmas)/ xmas
        xmas_return = (xmas_g + 1).product(axis = 0)
        xmas_return_df[str(range(2005,2017)[i]) + ' Xmas'] = xmas_return
    if  with_year == True:   
        for i in range(2005,2017):
            xmas_return_df[i] = y_return_df[i]

    return xmas_return_df


def yearly_return():
    y_return_df = pd.DataFrame()
    for i in range(len(year()[0])):
        start = year()[0][i]
        end = year()[1][i]
        y_data = close.ix[start: end]
        y_g = (y_data.shift(-1) - y_data)*1.0/y_data
        y_return = (y_g + 1).product(axis = 0)
        y_return_df[range(2005,2017)[i]] = y_return
    return y_return_df
    
def percentage_outperformer(which_holiday, return_list  = False):  #holiday == 'xmas', 'cheti', 'cny'
    y_return_df1 = yearly_return()
    
    xmas_return1 =  xmas_return(with_year = True)
    cheti_return1 = cheti_cny('cheti',False)
    cny_return1 = cheti_cny('cny', False)
    pct_list = []
    for col in range(11):
        year = range(2005,2017)[col]
        if which_holiday == 'xmas':
         
            colname = xmas_return1.columns[col]
            
            percentage = sum(1.0 * xmas_return1[colname] / y_return_df1[year] >1)*100.0/33
            print 'In %d year pre-Christmas period, %.2f%% STI stocks beat the year average return.' %(year, percentage)
            pct_list.append(percentage)

        if which_holiday == 'cheti':
            
            colname = cheti_return1.columns[col]
            percentage = sum(1.0 * cheti_return1[colname] / y_return_df1[year] >1)*100.0/33
            print 'In %d year pre Cheti Chand period, %.2f%% STI stocks beat the year average return.' %(year, percentage)
            pct_list.append(percentage)

        if which_holiday == 'cny':
            colname = cny_return1.columns[col]
            percentage = sum(1.0*cny_return1[colname] / y_return_df1[year] >1)*100.0/33
            print 'In %d year pre Chinese New Year period, %.2f%% STI stocks beat the year average return.' %(year, percentage)
            pct_list.append(percentage)

    if return_list == True:
        
        return pct_list

    
def which_stock_outperform(which_holiday, print_out = True):  # 'xmas' 'cheti' 'cny'
    outperformers = {}
    
    tickers = close.columns.values.tolist()
    xmas_return1 = xmas_return(True)
    cheti_return1 = cheti_cny('cheti',True)
    cny_return1 = cheti_cny('cny', True)
   
    for i, year in enumerate(range(2005,2017)):
        ticker_list = []
        for ticker in tickers:
            if which_holiday == 'xmas':
                try:    
                    normal = xmas_return1.ix[ticker, year]
                    xmas = xmas_return1.ix[ticker, str(year)+' Xmas']
                    judge = 1.0*xmas / normal >1
                    if judge:
                        if print_out == True:
                            print 'In year %d Christmas, stock %s beat the market'%(year, ticker)
                        ticker_list.append(ticker)
                except Exception as e:
                    print e
                outperformers[year] = ticker_list
            if which_holiday == 'cheti':
                try:    
                    normal = cheti_return1.ix[ticker, year]
                    colname = cheti_return1.columns[i]
                    cheti = cheti_return1.ix[ticker, colname]
                    judge = 1.0 * cheti / normal >1
                    if judge:
                        if print_out == True:
                            print 'In year %d Cheti Chand, stock %s beat the market'%(year, ticker)
                        ticker_list.append(ticker)
                except Exception as e:
                    print e
                outperformers[year] = ticker_list
            if which_holiday == 'cny':
                try:    
                    normal = cny_return1.ix[ticker, year]
                    colname = cny_return1.columns[i]
                    cny = cny_return1.ix[ticker,colname]
                    judge = 1.0*cny / normal >1
                    if judge:
                        if print_out == True:
                            print 'In year %d Chinese New Year, stock %s beat the market'%(year, ticker)
                        ticker_list.append(ticker)
                except Exception as e:
                    print e
                outperformers[year] = ticker_list
                
    return outperformers


def outperform_analysis(which_holiday, return_out = True, print_out = True):    
    out_dic = which_stock_outperform(which_holiday, print_out = False)
    good_tickers = []

    for key, item in out_dic.items():
        #print key, item
        good_tickers.append(item)
    #print good_tickers
    out_performers = [item for sublist in good_tickers for item in sublist]
    count = collections.Counter(out_performers)
    common = count.most_common()

    for item in common:
        if item[1] >= 4:
            if print_out == True:
                print 'Stock %s outperforms %d times out of 12 years!'%(item[0], item[1])
    if return_out == True:
        return common
        
def cheti_cny(which_holiday, with_year = True):  # holiday = cheti or cny
    
    return_df = pd.DataFrame()
    y_return_df = yearly_return()
    if which_holiday == 'cheti':
        holidays = chetis
    else: 
        holidays = cnys
        
    for holiday in holidays:
        start = strip_days(holiday, 10, goback =True)
        end = holiday
        cheti_cny = close.ix[start: end]
        cheti_cny_g = (cheti_cny.shift(-1) - cheti_cny)/ cheti_cny
        
        cheti_cny_return = (cheti_cny_g + 1).product(axis = 0)
        return_df[holiday + ' ' + which_holiday] = cheti_cny_return
        
    if  with_year == True:   
        for j in range(2005,2017):
            return_df[j] = y_return_df[j]

    return return_df

def mean_std(which_holiday): # return df
    table = {'xmas': xmas_return(False), 'cheti': cheti_cny('cheti', with_year = False),
            'cny': cheti_cny('cny', with_year = False), 'general': yearly_return()}
    data = table[which_holiday]
    mean = pd.DataFrame(ma.masked_invalid(data)).mean()
    std = pd.DataFrame(ma.masked_invalid(data)).std()
    #y_return = yearly_return().mean()
    mean_std = pd.concat([mean,std],axis = 1)
    mean_std.columns = ['Mean','Std']
    mean_std.index = data.mean().index
    return mean_std
        
def mean_std_all():
    mean3std = pd.DataFrame()
    #yearly_return().mean()mean_std()  # fix later
    mean3std = pd.concat([mean_std('xmas'),mean_std('cny'),mean_std('cheti'),mean_std('general')],axis =0)
    mean = {}
    std = {}
    for i, index in enumerate(mean3std.index):
        if i <= 11:
            mean['Xmas'] = mean3std.ix['2005 Xmas':'2016 Xmas','Mean'].mean()
            std['Xmas'] = mean3std.ix['2005 Xmas':'2016 Xmas','Std'].mean()
        elif i <= 22:
            mean['CNY'] = mean3std.ix['2005-02-08 cny':'2016-02-06 cny','Mean'].mean()
            std['CNY'] = mean3std.ix['2005-02-08 cny':'2016-02-06 cny','Std'].mean()
        elif i <= 33:
            mean['ChetiChand'] = mean3std.ix['2005-01-10 cheti':'2016-04-09 cheti','Mean'].mean()
            std['ChetiChand'] = mean3std.ix['2005-01-10 cheti':'2016-04-09 cheti','Std'].mean()
        else:
            mean['General year'] = mean3std.iloc[36]['Mean'].mean()
            std['General year'] = mean3std.iloc[36]['Std'].mean()
    print '=============== Mean ==============='
    for key, item in mean.items():
        print 'In %s mean return is %.4f'%(key, item)

    print '=============== Stdev ==============='    
    for key, item in std.items():
        print 'In %s mean standard deviation is %.4f'%(key, item)

def autocorrelation(ticker):
    data = close[ticker]
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(data, lags=40, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(data, lags=40, ax=ax2)
    plt.xlabel(ticker)
    plt.show()

def heatmap():
    corr_matrix = close.pct_change().dropna().corr()
    data1 = corr_matrix.values  # turn into an array matix
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    heatmap = ax1.pcolor(data1, cmap=plt.cm.spring)
    fig1.colorbar(heatmap)
    ax1.set_xticks(np.arange(data1.shape[1]) + 0.5, minor=False)
    ax1.set_yticks(np.arange(data1.shape[0]) + 0.5, minor=False)
    ax1.invert_yaxis()
    ax1.xaxis.tick_top()
    column_labels = corr_matrix.columns
    #row_labels = corr_matrix.index
    ax1.set_xticklabels(column_labels)
    ax1.set_yticklabels(column_labels)

    plt.xticks(rotation=90)
    heatmap.set_clim(-1,1)
    plt.tight_layout()
    #plt.savefig("correlations.png")
    #pickle.dump(ax1, file('gdp_corr.pickle', 'w'))
    plt.show()

def set_dummy(df, before = True):
    df_copy = df.copy()
    df_copy['xmas_dummy'] = 0
    df_copy['cheti_dummy'] = 0
    df_copy['cny_dummy'] = 0
    
    for i in range(12):
        # set xmas dummy
        startx = xmas_time()[0][i]
        endx = xmas_time()[1][i]
        print 'xmas start:', startx, 'xmas end:',endx
        if before == False:
            startx = endx
            endx = strip_days(startx, 10)
            print 'after xmas start:', startx, 'after xmas ends:', endx
        df_copy.ix[startx: endx,'xmas_dummy'] = 1
        print 'Year %d xmas dummy set!' %range(2005,2017)[i]
        
        # set cheti dummy
        startx = chetis[i]
        endx = strip_days(startx, 10)
        print 'cheti start:' ,startx,'cheti end:', endx
        if before == False:
            startx = endx
            endx = strip_days(startx, 10)
            print 'after cheti start:', startx, 'after cheti ends:', endx
        df_copy.ix[startx: endx,'cheti_dummy'] = 1
        print 'Year %d cheti dummy set!' %range(2005,2017)[i]
        
        # set cny dummy        
        startx = cnys[i]
        endx = strip_days(startx, 10)
        print 'cny start:' ,startx, 'cny ends:',endx
        if before == False:
            startx = endx
            endx = strip_days(startx, 10)
            print 'after cny start:', startx, 'after cny ends:', endx
        df_copy.ix[startx: endx,'cny_dummy'] = 1
        
        print 'Year %d cny dummy set!'%range(2005,2017)[i]
        print '============================================'
    return df_copy

# def after_holiday_dummy(df, which_holiday):
#     df_copy = df.copy()
#     df_copy['dummy'] = 0
#     for i in range(12):
#         if which_holiday == 'xmas':
#             startx = xmas_time()[0][i]          
#         if which_holiday == 'cheti':
#             startx = chetis[i]            
#         if which_holiday == 'cny':
#             startx = cnys[i]            
#         endx = strip_days(startx, 10)
#         df_copy.ix[startx: endx]['dummy'] = 1
#     return df_copy

def dummy_regr(df, ticker, which_dummy , three_together = False, log = False, GARCH = False):   # should fix this regression later!
    y = df[ticker]
    x = df[which_dummy]
    price_2 = (y ** 2).shift(-1)[:-1]
    if log ==True:
        y = np.log(y)
    if three_together == True:
        x = np.column_stack((df.xmas_dummy, df.cheti_dummy, df.cny_dummy))
    if GARCH == True:
        x = x[:-1]
        x = np.column_stack((x, price_2))
        y = (y ** 2)[:-1]
    model = sm.OLS(y, x)
    results = model.fit()
    return results.summary()







```
